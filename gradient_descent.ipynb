```python
import numpy as np
import matplotlib.pyplot as plt

# --- Target function ---
def func(x):
    return np.sin(x) + 0.5 * x**2   # You can change this

def dfunc(x):
    return np.cos(x) + x

# --- Gradient Descent ---
def gradient_descent(start, derivative_func, lr, steps):
    path = [start]
    x = start
    for _ in range(steps):
        x -= lr * derivative_func(x)
        path.append(x)
    return np.array(path)

# --- Main ---
start_point = 4
learning_rate = 0.5
steps = 8

x_path = gradient_descent(start_point, dfunc, learning_rate, steps)
print("x values during gradient descent:")
print(x_path)

# --- Plotting ---
x_plot = np.linspace(-6, 6, 400)
plt.plot(x_plot, func(x_plot), label='Target function')
plt.plot(x_path, func(x_path), label=f'Gradient Descent Path (lr={learning_rate})')
plt.scatter(x_path, func(x_path), color='black', zorder=5)

plt.title("Gradient Descent Visualization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

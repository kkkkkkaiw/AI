```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten
import matplotlib.pyplot as plt
import numpy as np
```

```python
# --- Load MNIST dataset ---
mnist = keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
# --- Inspect dataset ---
print(type(train_images))
print(train_images.shape)
print(type(train_labels))
print(train_labels.shape)
print(test_images.shape)
print(test_labels.shape)
```

```python
# --- Display 5 random images from training set ---
num_samples = 5
random_indices = np.random.choice(len(train_images), num_samples, replace=False)

for idx in random_indices:
    print(f"Image index: {idx}, Label: {train_labels[idx]}")
    plt.imshow(train_images[idx], cmap='binary')
    plt.title(f"Label: {train_labels[idx]}")
    plt.show()
```

```python
# --- Build the model ---
model = Sequential()
model.add(Flatten(input_shape=(28,28)))       # specify input shape
model.add(Dense(32, activation='relu'))       # hidden layer
model.add(Dense(10, activation='sigmoid'))    # output layer
```

```python
# --- Compile model ---
model.compile(
    optimizer='rmsprop',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# --- Train model ---
model.fit(
    train_images, train_labels,
    epochs=15,
    batch_size=512,
    verbose=1
)
# --- Evaluate model ---
test_loss, test_accuracy = model.evaluate(test_images, test_labels)
print(f"Test accuracy: {test_accuracy:.4f}")
```
